```python
import csv
import folium
import timeit
import numpy as np
import pandas as pd
from folium.plugins import FastMarkerCluster
from datetime import datetime
from time import process_time
```

```python
with open('../data/core_permits.csv',encoding='ascii',errors='ignore') as csvfile:
    permits = pd.read_csv(csvfile)
#  Change data types to reflect actual data types   
permits['Issue Date']= pd.to_datetime(permits['Issue Date'], format='%Y-%m-%d')
permits['Final Date']= pd.to_datetime(permits['Final Date'], format='%Y-%m-%d')
permits['Expiration Date']= pd.to_datetime(permits['Expiration Date'], format='%Y-%m-%d')
permits['Application Date']= pd.to_datetime(permits['Application Date'], format='%Y-%m-%d')
permits['Accident_Count']= np.random.randint(low = 1, high = 6, size = permits.shape[0])

commercial_permits = permits[permits['Category'] == 'COMMERCIAL']
```

```python
lat = np.mean(commercial_permits['Latitude'])
lon = np.mean(commercial_permits['Longitude'])
folium.Map(location=[lat, lon], zoom_start = 11)
```

```python
def create_map(data):
    my_map = folium.Map(location=[np.mean(data['Latitude']), np.mean(data['Longitude'])], zoom_start = 11)
    for index, row in data.iterrows():
        folium.CircleMarker(
            location = [row['Latitude'], row['Longitude']],
            radius=row['Accident_Count'],
            fill = True,
            popup = folium.Popup(str("Address: " + row['Address'])),
            fill_color='#132b5e'
        ).add_to(my_map)
    return my_map
```

```python
t = process_time()
create_map(commercial_permits)
```

```python
elapsed_time = process_time() - t
elapsed_time
```
